## Cleaning

```python
#extracting the exact genre from the text in genre column
df_movies['genre_n'] = ""
```

```python
list_of_genres = []
seperator = ', '
for ind in df_movies.index:
    genres_list_json_parsed = json.loads(df_movies['genres'][ind])
    updated_genres_list = list(map(lambda d: d['name'], genres_list_json_parsed))
    stringified_list = seperator.join(updated_genres_list)
    list_of_genres.append(stringified_list)

```

```python
df_movies['genre_n'] = list_of_genres
```

```python
df_movies.head()
```

```python
df_movies.info()
```

```python
# Changing release_date to datetime.
df_movies['release_date'] = pd.to_datetime(df_movies['release_date'])
```

```python
## Dropping homepage column.keywords,original_title,overview,tagline
df_movies.drop(['homepage','keywords','original_title','overview','tagline'], axis=1,inplace=True)
#df_movies.drop(['genres'],axis=1,inplace = True)
```

```python
#id datatype is int64 instead of string
df_movies['id']=df_movies['id'].astype(str)
```

```python
df_movies.head()
```

```python
# extracting the exact companies from the text in production_companies column
list_of_companies = []
seperator = ', '
for ind in df_movies.index:
    companies_list_json_parsed = json.loads(df_movies['production_companies'][ind])
    updated_companies_list = list(map(lambda d: d['name'], companies_list_json_parsed))
    stringified_list = seperator.join(updated_companies_list)
    list_of_companies.append(stringified_list)
```

```python
df_movies['production_companies_n']=""
```

```python
df_movies['production_companies_n'] = list_of_companies
```

```python
df_movies.drop(['production_companies'],axis=1,inplace = True)
```

```python
#extracting the exact countries from the text in production_countries column
list_of_countries = []
seperator = ', '
for ind in df_movies.index:
    countries_list_json_parsed = json.loads(df_movies['production_countries'][ind])
    updated_countries_list = list(map(lambda d: d['name'], countries_list_json_parsed))
    stringified_list = seperator.join(updated_countries_list)
    list_of_countries.append(stringified_list)
```

```python
df_movies['production_countries_n']=""
```

```python
df_movies['production_countries_n'] = list_of_countries
```

```python
df_movies.drop(['production_countries'],axis=1,inplace = True)
```

```python
#extracting languages from the test in spoken_languages column.
list_of_languages = []
seperator = ', '
for ind in df_movies.index:
    languages_list_json_parsed = json.loads(df_movies['spoken_languages'][ind])
    updated_languages_list = list(map(lambda d: d['name'], languages_list_json_parsed))
    stringified_list = seperator.join(updated_languages_list)
    list_of_languages.append(stringified_list)
```

```python
df_movies['spoken_languages_n']=""
```

```python
df_movies['spoken_languages_n']= list_of_languages
```

```python
df_movies.drop(['spoken_languages'],axis=1,inplace = True)
```
